```python
import os
import pandas as pd
Dataset1 = pd.read_excel('EXTENDED1.xlsx')
Dataset2 = pd.read_excel('EXTENDED2.xlsx')
MainDataset = pd.merge(Dataset1, Dataset2, how='left',on = 'ID')
Dataset3 = pd.read_excel('EXTENDED3.xlsx')

MainDataset = pd.merge(MainDataset, Dataset3, how='left',on = 'ID')
MainDataset
```

```python
MainDataset.to_csv("output.csv", index=False) 
Dataset = pd.read_csv('output.csv')
columns = [
"Do you need to win in order to derive enjoyment from games and sports?",
"Do you often try to do more than one thing at a time?",
"Do you feel confident when you start any project?",
"Do you get angry instantly if anything goes wrong?",
"Do you find it difficult and frustrating to wait in line?",
"Are you eager to harm yourself to defend your beliefs?", 
"Are you impatient in general?",
"CGPA",
"Gender",

]

# Get names of indexes for which column Age has value 30

Dataset = Dataset[columns].copy()

Dataset = Dataset.replace(['Yes', 'No', 'Maybe','Female','Male'], 
                     [1, 2, 3,1,2]) 
Dataset = Dataset.fillna(0)
indexNames = Dataset[ Dataset['CGPA'] == 0 ].index
Dataset.drop(indexNames,inplace = True)
indexNames = Dataset[ Dataset['Are you impatient in general?'] == 0 ].index
Dataset.drop(indexNames,inplace = True)
Dataset.head()
```

```python
Dataset2 = Dataset.iloc[:,0:6].div(Dataset.CGPA, axis=0) 

col_list= list(Dataset2)
A = Dataset[col_list].sum(axis=1)
Dataset2['Are you impatient in general?'] =Dataset.loc[:,'Are you impatient in general?'].mul(A,axis=0)
Dataset2['Gender'] = Dataset['Gender'].values
Dataset2['CGPA'] = Dataset['CGPA'].values

#Dataset2['Are you impatient in general?'] = Dataset['Are you impatient in general?'].values
```
